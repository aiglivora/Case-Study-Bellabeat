# Case Study Bellabeat

## Résumé
Bellabeat est une entreprise innovante spécialisée dans les produits de santé connectés destinés aux femmes. ### Analyse des activité diurnes
Afin d'analyser les activités diurnes, les tables dailyActivityMars et dailyActivityAvr seront fusionnées.

```r
dailyActivity <- rbind(dailyActivityMars, dailyActivityAvr)
```

Selon "U.S. Department of Health and Human Services. Physical Activity Guidelines for Americans, 2nd edition. Washington, DC: U.S.Department of Health and Human Services; 2018", il est recommandé aux adultes de pratiquer entre 150 et 300 minutes par semaine d'activité physique aérobie d'intensité modérée, ou entre 75 et 150 minutes par semaine d'activité aérobie d'intensité vigoureuse. Par ailleurs, une recommandation populaire, valable pour tous les âges, préconise de marcher 10 000 pas par jour. Dans ce cadre, notre analyse portera sur le temps alloué à l'activité physique par semaine et sur le nombre de pas réalisés par jours des utilisateurs.

#### Analyse du temps consacré à l'activité physique
Dans le cadre de l'analyse du temps consacré à l'activité physique, nous examinerons trois catégories : moins de 150 minutes par semaine, entre 150 et 300 minutes par semaine, et plus de 300 minutes par semaine. Pour cette analyse, plusieurs modifications seront apportées à la table dailyActivity :

-Ajout d'une colonne *dayWeek* indiquant le jour de la semaine pour chaque entrée.

-Ajout d'une colonne *week* identifiant la semaine correspondante.

-Ajout d'une colonne *totalActivity*, calculée comme la somme des valeurs VeryActiveMinutes et FairlyActiveMinutes par jour.

```r
dailyActivity$dayWeek <- wday(dailyActivity$ActivityDate, label = TRUE, abbr = FALSE)

dailyActivity <- dailyActivity %>%
  mutate(week = as.integer(format(ActivityDate, "%U")))%>%
  mutate(totalActivity = VeryActiveMinutes + FairlyActiveMinutes)

df_status(dailyActivity)
```

Ensuite, les utilisateurs seront classés dans des zones d'activité spécifiques, et le nombre d'ID par zone sera comptabilisé chaque semaine, notamment pour les semaines 11 à 19, où chaque semaine comporte 7 jours complets. 

```r
activity_per_week <- dailyActivity %>%
  group_by(Id, week) %>%
  summarise(total_activity_week = sum(totalActivity), .groups = 'drop') %>% 
  mutate(
    Zone = case_when(
     total_activity_week < 150 ~ "Moins de 150 minutes",
     total_activity_week >= 150 & total_activity_week <= 300 ~ "150 à 300 minutes",
     total_activity_week > 300 ~ "Plus de 300 minutes"
     )
   )
```

```r
activity_zone_summary <- activity_per_week %>%
  filter(week > 10 & week <= 19) %>% 
  group_by(week, Zone) %>%
  summarise(nb_Id = n_distinct(Id), .groups = "drop") %>%   # Nombre d'ID par zone chaque semaine
  group_by(week) %>%
  mutate(
    percentage = nb_Id / sum(nb_Id) * 100,
    Zone = factor(Zone, levels = c("Plus de 300 minutes", 
                                   "150 à 300 minutes", 
                                   "Moins de 150 minutes"))     
  )
head(activity_zone_summary)
```

La figure ci-dessous présente le pourcentage d'utilisateurs dans chaque catégorie pour les semaines 11 à 19. En raison d'un nombre inférieur à cinq utilisateurs portant la montre lors des semaines 11 et 12, les résultats de ces semaines ne sont pas représentatifs. Pour les semaines 13 à 19, les résultats montrent que de 35 % à 78 % des utilisateurs pratiquent moins de 150 minutes d'activité physique par semaine, comme recommandé.


```r
ggplot(activity_zone_summary, aes(x = factor(week), y = nb_Id, fill = Zone)) +
  geom_bar(stat = "identity", position = "stack") +
  geom_text(aes(label = paste0(round(percentage, 1), "%")), 
            position = position_stack(vjust = 0.5),
            color = "white", 
            size = 4) + 
  labs(
    title = "Répartition du temps d'activité physique des utilisateurs par zone d'intensité",
    x = "Semaine",
    y = "Nombre d'itilisateur",
    fill = "Zone d'intensité"
  ) +
  theme_minimal()
```

#### Analyse du nombre de pas par jour
Une analyse approfondie du nombre de pas quotidiens réalisés par les utilisateurs a été menée. Cette analyse se concentre sur le nombre de jours où les utilisateurs ont enregistré,selon les trois catégories suivantes :

-Moins de 5 000 pas

_Entre 5 000 et 10 000 pas

-Plus de 10 000 pas.

```r
activity_date<- dailyActivity %>%
  mutate(
    zoneSteps = case_when(
      TotalSteps < 5000 ~ "moins 5K pas",
      TotalSteps >= 5000 & TotalSteps < 10000 ~ "5K à 10k pas",
      TotalSteps >= 10000 ~ "10K pas et plus"
    )
  ) %>% 
  group_by(Id, zoneSteps) %>%
  summarise(n_daySteps = n(), .groups = 'drop') %>%
  mutate(
    zoneSteps = factor(zoneSteps, levels = c("10K pas et plus", "5K à 10k pas", "moins 5K pas"))
  )
head(activity_date)
```

```r
avg_days_less_5K <- dailyActivity %>%
  filter(TotalSteps < 5000) %>% 
  group_by(Id) %>%                
  summarise(n_days_less_5K = n(), .groups = 'drop') %>%  
  summarise(mean_days_less_5K = mean(n_days_less_5K)) %>% 
  pull(mean_days_less_5K)
cat("Moyenne de jours avec moins de 5000 pas :", round(avg_days_less_5K, 1), "\n")
```

La figure ci-dessous présente la répartition du nombre de jours de marche par utilisateur selon les trois catégories.Les résultats montrent qu'aucun utilisateur n'a atteint la recommandation des 10 000 pas par jour sur l'ensemble des jours observés.Par ailleurs, tous les utilisateurs enregistrent au moins un jour avec moins de 5 000 pas, un niveau correspondant à la moyenne quotidienne de pas sans activité physique significative.La moyenne de jours avec moins de 5000 pas est de 10 jours.

```r
ggplot(activity_date, aes(x = factor(Id), y = n_daySteps, fill = zoneSteps)) + 
  geom_bar(stat = "identity", , position = "stack") + 
  geom_text(aes(label = n_daySteps),
            position = position_stack(vjust = 0.5),  
            angle = 0,
            color = "black", 
            size = 3) + 
  labs(title = "Répartition du nombre de jours de marche par zone pour chaque utilisateur",
       x = "ID", 
       y = "Nombre de jours",
       fill = "Zone d'activité") +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 90, hjust = 1))
```

### Analyse des activités noctures
Selon "Sante publique France" les besoins de sommeil varient d’une personne à l’autre et selon l’âge. La durée idéale d'une nuit de sommeil est celle qui permet de se réveiller en se sentant reposé et performant. En moyenne, un adulte dort entre 7 et 8 heures par nuit, bien que certains individus, appelés « courts dormeurs », se contentent de 3 heures 30, tandis que d'autres, appelés « longs dormeurs », peuvent dormir entre 11 et 12 heures par nuit. Dans cette étude, les utilisateurs sont classés en quatre catégories afin d'analyser leur comportement de sommeil en fonction des données recueillies par la montre connectée. Les catégories sont les suivantes :

-Moins de 7 heures de sommeil

-Entre 7 et 8 heures de sommeil

-Entre 8 et 10 heures de sommeil

-Plus de 10 heures de sommeil 

```r
sleep_time<- sleepAvr %>%
  mutate(
    zone_sleep = case_when(
      TotalMinutesAsleep < 420 ~ "moins 7H",
      TotalMinutesAsleep >= 420 & TotalMinutesAsleep <= 480 ~ "7H à 8H",
      TotalMinutesAsleep > 480 & TotalMinutesAsleep <= 600 ~ "8H à 10H",
      TotalMinutesAsleep > 600 ~ "plus 10H"
    )
  ) %>% 
  group_by(Id, zone_sleep) %>%
  summarise(n_daySleep = n(), .groups = 'drop') %>%  
  mutate(
    zone_sleep = factor(zone_sleep, levels = c("plus 10H", "8H à 10H", "7H à 8H", "moins 7H")) 
  )
head(sleep_time)
```

La figure ci-dessous illustre la répartition du nombre de jours de sommeil par catégorie pour chaque utilisateur. La durée de sommeil varie d'un jour à l'autre, et n'est pas régulière. Pour la majorité des utilisateurs, la durée de sommeil se situe entre moins de 7 heures et 10 heures par nuit. Certains utilisateurs dorment plus de 10 heures par nuit.

```r
ggplot(sleep_time, aes(x = factor(Id), y = n_daySleep, fill = zone_sleep)) + 
  geom_bar(stat = "identity", , position = "stack") + 
  geom_text(aes(label = n_daySleep),
            position = position_stack(vjust = 0.5),
            angle = 0,
            color = "black", 
            size = 3) + 
  labs(title = "Répartition du nombre de jours de sommeil par zone et par utilisateur",
       x = "ID", 
       y = "Nombre de jours",
       fill = "catégorie") +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 90, hjust = 1))
```

## Conclusion
Les données du tracker d’activité Fitbit montrent l’activité quotidienne d’une trentaine de personnes sur une période de deux mois (mars et avril 2016). L’analyse du nombre de jours de port de la montre connectée indique que les utilisateurs ont tendance à la porter de manière plus régulière pendant leur sommeil que durant la journée. Cependant, avec seulement deux mois de données, cette tendance ne peut pas être confirmée de manière concluante.

L’analyse des activités des utilisateurs révèle que plusieurs d’entre eux ne respectent pas les recommandations de pratiquer au moins 150 minutes par semaine d’activité physique aérobie d’intensité modérée, ni de marcher au moins 10 000 pas par jour. En ce qui concerne le sommeil, la durée varie d’un jour à l’autre et n’est pas régulière. Pour la majorité des utilisateurs, la durée de sommeil se situe entre moins de 7 heures jusqu'à 10 heures par nuit, tandis que certains dorment plus de 10 heures par nuit.

Cependant, les données sur les activités diurnes et nocturnes ne fournissent pas d’informations suffisamment détaillées pour évaluer précisément l’état de santé des utilisateurs.

Afin de renforcer la croissance et la visibilité de Bellabeat sur le marché des appareils de santé connectés, plusieurs actions pourraient être envisagées :

1. Organiser des concours pour encourager les utilisateurs à atteindre les objectifs d’activité physique quotidiens et hebdomadaires recommandés, en fonction de groupes d’âge.

2. Améliorer le design de la montre pour la rendre plus confortable à porter durant la journée, tout en l’adaptant aux vêtements quotidiens, et également pour garantir qu’elle ne gêne pas le sommeil.   

3. Développer des algorithmes permettant aux utilisateurs de connaître l'état de leur corps afin de recommander l'intensité d'activité physique la plus adaptée, tout en leur offrant la possibilité de suivre et d’évaluer la qualité de leur sommeil ainsi que les différentes phases du sommeil.
